Каково среднее для всех немошеннических операций при пересчете в доллары США?

```python
exchange_df = pd.read_parquet('historical_currency_exchange.parquet')
exchange_df.head()
```

```python
exchange_df.info()
```

```python
exchange_df['date'] = pd.to_datetime(exchange_df['date']).dt.date
```

```python
isnt_fraud_df = df[df['is_fraud'] == False].copy()
isnt_fraud_df['date'] = isnt_fraud_df['timestamp'].dt.date
```

```python
merged = isnt_fraud_df.merge(exchange_df, how='left', on='date')
merged
```

```python
def to_usd(row):
    currency = row['currency']
    amount = row['amount']
    rate = row.get(currency)
    if currency == 'USD':
        return amount
    return (amount/rate)

merged['in_usd']=merged.apply(to_usd, axis=1)
merged
```

```python
df['currency'].value_counts()
```

```python
avg_isnt_fraud_usd = merged['in_usd'].mean()
avg_isnt_fraud_usd
```

Каково среднеквадратичное отклонение среди всех немошеннических операций при пересчете в доллары США?

```python
avg_isnt_fraud_usd_std = merged['in_usd'].std()
avg_isnt_fraud_usd_std
```

Определите, сколько клиентов демонстрируют потенциально опасное поведение: для каждого клиента возьмите медианное значение числа уникальных продавцов из показателя активности «за последний час» на момент транзакции и посчитайте, сколько клиентов строго превышают 95-ый квантиль от этого показателя.
В ответе напишите единственное целое число — количество таких клиентов.

```python
df['unique_merchants'] = df['last_hour_activity'].apply(lambda x: x['unique_merchants'] if pd.notnull(x) else pd.NA)
df.head()
```

```python
median_merchant_per_client = df.groupby('customer_id')['unique_merchants'].median()
median_merchant_per_client
```

```python
quantile_95 = median_merchant_per_client.quantile(0.95)
quantile_95
```

```python
high_risk_customers = (median_merchant_per_client > quantile_95).sum()
high_risk_customers
```
